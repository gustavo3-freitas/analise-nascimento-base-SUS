#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import PyPDF2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
df = pd.read_csv('SINASC_RO_2019.csv')

df.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[(df['APGAR5'] >= 0) & (df['APGAR5'] <= 3),  'APGAR5_CAT'] = 'asfixia grave'
df.loc[(df['APGAR5'] >= 4) & (df['APGAR5'] <= 5),  'APGAR5_CAT'] = 'asfixia moderada'
df.loc[(df['APGAR5'] >= 6) & (df['APGAR5'] <= 7),  'APGAR5_CAT'] = 'asfixia leve'
df.loc[(df['APGAR5'] >= 8) & (df['APGAR5'] <= 10), 'APGAR5_CAT'] = 'normal'

df.head()

print(df.columns)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
plt.figure(figsize=(10, 6))
plt.tight_layout()

sns.boxplot (x='GESTACAO', y='APGAR5', hue='SEXO', data=df, )
plt.xticks(rotation=45)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tab_cruzada = pd.crosstab(df['GESTACAO'], df['APGAR5'])
tab_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
pctlin = tab_cruzada.div(tab_cruzada.sum(axis=1), axis = 0)
print(pctlin)
```
